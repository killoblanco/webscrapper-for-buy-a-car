```python
import os
import pandas as pd

df = pd.read_csv(os.path.join('..', 'data', 'data.csv'))
df.describe()
```

```python
df.dtypes
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
df = df.dropna(axis=0, how='any')
```

```python
df.groupby('brand').count()
```

```python
kicks = df[df['brand'].str.contains('kicks', case=False)]
kicks.dtypes
```

```python
price = pd.to_numeric(kicks.loc[:, 'price'])
km = pd.to_numeric(kicks.loc[:, 'km'])
year = pd.to_numeric(kicks.loc[:, 'year'])

kicks.loc[:, 'price'] = price
kicks.loc[:, 'km'] = km
kicks.loc[:, 'year'] = year

kicks.dtypes
```

```python
k_exclusive = kicks[kicks['model'].str.contains('exclusive', case=False)]
```
